```python
with  open('smartprix.html','r',encoding='utf-8') as f:
  html = f.read()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
name = []
price = []
rating = []
sim = []
processor = []
ram = []
battery = []
display = []
camera = []
card = []
os = []




for i in soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'}):
  name.append(i.find('h2').text)
  price.append(i.find('span',{'class':"price"}).text)

  try:
    rating.append(i.find('div',{'class':'score rank-1-bg'}).find('b').text)
  except:
    rating.append(np.nan)




  x = i.find('ul',{'class':'sm-feat specs'}).find_all('li')
  try:
    sim.append((x[0]).text)
  except:
    sim.append(np.nan)

  try:
    processor.append((x[1]).text)
  except:
    processor.append(np.nan)

  try:
    ram.append((x[2]).text)
  except:
    ram.append(np.nan)

  try:
    battery.append((x[3]).text)
  except:
    battery.append(np.nan)

  try:
    display.append((x[4]).text)
  except:
    display.append(np.nan)

  try:
    camera.append((x[5]).text)
  except:
    camera.append(np.nan)

  try:
    card.append((x[6]).text)
  except:
    card.append(np.nan)

  try:
    os.append((x[7]).text)
  except:
    os.append(np.nan)
```

```python
df = pd.DataFrame({
    'model': name,
    "price":price,
    "rating":rating,
    'sim':sim,
    'processor':processor,
    'ram':ram,
    'battery':battery,
    'display':display,
    'camera':camera,
    'card':card,
    'os':os


})
```

```python
df.to_csv('Smartphone.csv')
```
